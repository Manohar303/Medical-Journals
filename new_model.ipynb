```python
parsed_pdf = parser.from_file("888-97.pdf") 
data = parsed_pdf['content']  
text=data.strip()
text =text.split('Reference')[0]
text=text.replace("-\n","")
text="\n".join(text.split(","))
```

```python
text=re.sub("[^A-Za-z-0-9+–-" "]+"," ",text)
#text=re.sub(r'([^\w\s]|_)+(?=\s|$)', '', text)
text=re.sub(r'\b[0-9]+\b', '', text)
```

```python
text = re.sub("@\S+", " ", text)
text = re.sub("https*\S+", " ", text)
text = re.sub("#\S+", " ", text)
text = re.sub("\'\w+", '', text)
text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
text = re.sub('\s{2,}', " ", text)
text
```

```python
text = text.encode('ascii', 'ignore').decode()#to remove unicode char
text
```

```python
length=len(text)
avg_length=int(length/3)
avg_length
```
